задача 1

```python
import string


def task1(inputString):
    chars = list(string.ascii_lowercase)
    charCounts = dict(zip(chars, [0]*len(chars)))
    for c in inputString:
        charCounts[c] += 1

    vals = list(charCounts.values())

    #subsCount = vals[-1] // 2
    subsCount = 0
    while(True):
        for i in range(len(vals)):
            vals[i] += subsCount
            subsCount = vals[i] // 2
            vals[i] = vals[i] % 2
        
        if(subsCount == 0):
            break
    
    resDict = dict(zip(chars, vals))
    resStr = []
    for (key, val) in zip(resDict.keys(), resDict.values()):
      assert(val < 2)
      if(val == 1):
          resStr.append(key)

    return ''.join(resStr)
```

```python
tests = ['zzzab', 'sberbank', 'zzzabb', 'czz', 'test', 'queue', 'xxyz']
```

```python
for test in tests:
    print(f'{test} ---> {task1(test)}')
```

```python
import math
```

```python
def decomp(n):
    multipliers = [] # первое число - 2, оно простое
    powers = [1] * (n - 1) # степени разложения факториала, изначально это числа 2,3,4,5,...,n
    for i in range(n - 1):
        num = i + 2
        isPrime = True
        for j in range(2, math.floor(math.sqrt(num)) + 1):
            if(num % j == 0):
                multipliers.append([j] + multipliers[(num // j) - 2])
                isPrime = False
                break

        if(isPrime):
            multipliers.append([num])

        powers[i] -= 1
        for mult in multipliers[-1]:
            powers[mult - 2] += 1

    return ''.join(f'{"" if x == 0 else f"{i+2} * " if x == 1 else f"{i+2}^{x} * "}' for (i,x) in enumerate(powers))[:-3]
```

```python
tests = [12, 22, 25, 4000]
for test in tests:
    print(f'{test}! ---> {decomp(test)}')
```

```python
def parseNum(numbers):
    parseDict = {
        'один': 1,
        'одна': 1,
        'два': 2,
        'две': 2,
        'три': 3,
        'четыре': 4,
        'пять': 5,
        'шесть': 6,
        'семь': 7,
        'восемь': 8,
        'девять': 9,
        'десять': 10,
        'одиннадцать': 11,
        'двенадцать': 12,
        'тринадцать': 13,
        'четырнадцать': 14,
        'пятнадцать': 15,
        'шестнадцать': 16,
        'семнадцать': 17,
        'восемнадцать': 18,
        'девятнадцать': 19,
        'двадцать': 20,
        'тридцать': 30,
        'сорок': 40,
        'пятьдесят' : 50,
        'шестьдесят' : 60,
        'семьдесят' : 70,
        'восемьдесят' : 80,
        'девяносто': 90,
        'сто': 100,
        'двести': 200,
        'триста': 300,
        'четыреста': 400,
        'пятьсот': 500,
        'шестьсот': 600,
        'семьсот': 700,
        'восемьсот': 800,
        'девятьсот': 900
    }
    num = 0
    for n in numbers:
        num += parseDict[n]

    return num
```

```python
parseNum(['девятьсот', 'девяносто', 'девять'])
```

```python
def str2num(numString):

    if(numString == 'ноль'):
        return 0
    if(numString == 'один миллион'):
        return 1000000

    nums = numString.split()
    thousPos = -1
    for (i, n) in enumerate(nums):
        if('тысяч' in n):
            thousPos = i
            break

    if(thousPos > 0):
        thousands = parseNum(nums[:thousPos])*1000
    elif thousPos == 0:
        thousands = 1000
    else:
        thousands = 0
    
    if(thousPos < len(nums) - 1):
        hundreds = parseNum(nums[thousPos+1:])
    else:
        hundreds = 0
    
    return thousands + hundreds
```

```python
tests = [
         'девятьсот девяносто девять тысяч девятьсот девяносто девять',
         'ноль',
         'один миллион',
         'тысяча',
         'две тысячи',
         'тысяча триста',
         'тридцать пять',
         'тринадцать тысяч сорок один',
         'одна тысяча',
         'одна тысяча двадцать',
         'тысяча двадцать',
         'сто тысяч']
for test in tests:
    print(f'{test} ---> {str2num(test)}')
```
